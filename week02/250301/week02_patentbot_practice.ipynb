```python
from langchain.docstore.document import Document

split_docs = []

for page_doc in filtered_pages:
    # 페이지 텍스트를 줄 단위로 쪼갬
    lines = page_doc.page_content.split("\n")
    
    # 대략 절반 지점
    half_idx = len(lines) // 2
    
    # 상반(혹은 좌측 절반) 텍스트
    left_part = "\n".join(lines[:half_idx])
    # 하반(혹은 우측 절반) 텍스트
    right_part = "\n".join(lines[half_idx:])
    
    # 나뉜 두 개의 Document 생성
    doc_left = Document(
        page_content=left_part,
        metadata=page_doc.metadata  # 필요하면 컬럼 정보 등을 추가로 넣을 수도 있음
    )
    doc_right = Document(
        page_content=right_part,
        metadata=page_doc.metadata
    )
    
    split_docs.append(doc_left)
    split_docs.append(doc_right)
```

```python
import re

def remove_page_headers_and_footers(raw_text: str) -> str:
    lines = raw_text.split("\n")
    filtered_lines = []

    for line in lines:
        line_stripped = line.strip()
        
        # 1) 숫자만 있는 줄(페이지 번호 등) -> 제거
        if re.match(r'^\d+$', line_stripped):
            continue
        
        # 2) 특정 문구(지식재산권 용어 사전, 영한사전 등) -> 제거
        if "지식재산권 용어 사전" in line_stripped:
            continue
        if "영한사전" in line_stripped:
            continue
        
        # 3) 대문자 1~3글자만 있는 라인 -> 제거 (예: "A", "AA", "AB", "AAA" 등)
        #    (실제 문서 구조에 따라 조정 가능)
        if re.match(r'^[A-Z]{1,3}$', line_stripped):
            continue
        
        filtered_lines.append(line)
    
    clean_text = "\n".join(filtered_lines)
    return clean_text
```

```python
# 2) 헤더/푸터 제거
clean_text = remove_page_headers_and_footers(raw_text)
```

```python
import re

def split_dictionary_entries(text: str):
    """
    항목 패턴 예시:
    - 첫 줄: 영문 용어 (첫 글자가 알파벳)
    - 두 번째 줄: 한글 용어 (첫 글자가 한글)
    - 그 이후: 설명 (다음 줄이 영문으로 시작하기 전까지)
    """
    pattern = r'(?m)^(?P<eng>[A-Za-z][^\n]*)\n(?P<kor>[가-힣][^\n]*)(?P<desc>(?:\n(?![A-Za-z]).*)*)'
    
    entries = []
    for match in re.finditer(pattern, text):
        entry = match.group(0).strip()
        entries.append(entry)
    return entries
```

```python
# 항목별로 분리
entries = split_dictionary_entries(clean_text)

print("추출된 항목 수:", len(entries))
for i, entry in enumerate(entries):
    print(f"[Entry {i}]\n{entry}\n{'-'*50}")
```

```python
from datetime import datetime

# 오늘의 날짜를 한글 형식으로 포맷 (예: 2025년 03월 01일)
today_date = datetime.today().strftime("%Y년 %m월 %d일")
```

```python
prompt_template = """
너는 지식재산용어 챗봇 '지식재산봇'이야. 만약 누군가 네게 네가 누구냐고 물으면, 반드시 '지식재산봇'이라고 답변해.
참고 문서를 바탕으로 답변하고, 만약 참고 문서에 있는 내용이라면 답변 끝에 '검색 기반 답변이에요'라고 붙여.
참고 문서에 없는 내용이라면 '제 지식으로 기반한 답변이에요'라고 붙여서 답변하면 돼.
만약 질문이 "오늘이 무슨 날짜야?"와 같이 오늘의 날짜를 묻는 경우, 현재 날짜에 해당하는 "{today_date}"이라고 정확하게 답변해줘.

이전 대화:
{{chat_history}}

참고 문서: {{context}}

질문: {{question}}
답변:""".format(today_date=today_date)

# 이제 프롬프트 템플릿은 {today_date} 부분이 이미 값으로 대체되었으므로,
# input_variables에는 'today_date'가 포함되지 않아야 합니다.
qa_prompt = PromptTemplate(template=prompt_template, input_variables=["chat_history", "context", "question"])
```

```python
crossencoder = CrossEncoder('BAAI/bge-reranker-v2-m3')

# Faiss + Cross Encoder 기반 Retriever 함수 정의
def cross_encoder_retriever(question):
    """Faiss로 검색한 문서들을 Cross Encoder로 재정렬하여 최종 결과 반환"""

    # 질문으로부터 첫번째 검색 결과를 얻는다.
    retrieved_docs = retriever.invoke(question)  # Faiss 기반 검색
    retrieved_texts = [doc.page_content for doc in retrieved_docs]

    # Cross Encoder로 첫번째 검색 결과에 대해서 일일히 점수를 다시 부여한다.
    query_doc_pairs = [(question, doc) for doc in retrieved_texts]
    cross_scores = crossencoder.predict(query_doc_pairs)

    # 점수가 가장 높은 문서 2개만 남긴다.
    best_indices = np.argsort(cross_scores)[::-1][:2]
    best_docs = [retrieved_texts[i] for i in best_indices]

    separator = "\n" + ("---" * 50) + "\n"
    return separator.join(best_docs)
```

```python
# 최근 대화만 기억하도록 하는 함수
class LimitedChatMessageHistory(BaseChatMessageHistory):
    def __init__(self, max_conversations: int = 3):
        self.full_messages: List[BaseMessage] = []  # 전체 메시지 저장 리스트
        self.messages: List[BaseMessage] = []       # 제한된 메시지 저장 리스트
        self.max_messages = (max_conversations - 1) * 2  # 최근 대화 쌍(사용자+AI)만 저장

    def add_message(self, message: BaseMessage) -> None:
        self.full_messages.append(message)  # 전체 기록에 추가
        self.messages.append(message)       # 제한된 기록에도 추가
        if len(self.messages) > self.max_messages:
            self.messages.pop(0)  # 제한 초과 시 가장 오래된 메시지 제거

    def clear(self) -> None:
        self.messages = []  # 대화 기록 초기화

    def get_messages(self) -> List[BaseMessage]:
        # 최근 대화 쌍을 포함해 반환 (추가로 과거 대화가 있다면 그 일부 포함)
        return self.full_messages[-self.max_messages-2:]
```

```python
session_id = "test_00"  # 현재 세션 ID 지정
message_store = {session_id: LimitedChatMessageHistory(max_conversations=3)}  # 세션별 히스토리 저장소
```

```python
# chat_history를 출력하기 위한 함수
def format_history(chat_history):
    formatted = ""
    for msg in chat_history:
        if msg.type == "human":
            formatted += f"사용자: {msg.content}\n"
        elif msg.type == "ai":
            formatted += f"모두봇: {msg.content}\n"
    return formatted
```

### 멀티턴을 RAG에 적용

```python
def logging_retriever(query):
    docs = cross_encoder_retriever(query)
    print("검색된 문서들:", docs)  # 검색된 문서를 출력
    return docs
```

```python
# LangChain Chain 설정 (Cross Encoder 적용)
chain = (
    {
        # cross_encoder_retriever 사용하지 않을 때 -> 첫번째 검색기 넣기
        "context": itemgetter("question") | RunnableLambda(logging_retriever),
        "question": itemgetter("question"),
        "chat_history": itemgetter("chat_history") | RunnableLambda(format_history),
    }
    | qa_prompt
    | llm
    | StrOutputParser()
)
```

```python
chain_with_chat_history = RunnableWithMessageHistory(
    chain,
    lambda sid: message_store[sid],            # 세션 ID로 해당 채팅 히스토리 반환
    input_messages_key="question",             # 입력 데이터에서 질문을 추출하는 키
    history_messages_key="chat_history"        # 체인에 전달할 히스토리 키 이름
)
```

### 그라디오 실행

```python
with gr.Blocks() as demo:
    chatbot = gr.Chatbot(label="지식재산봇")              # 챗봇 대화 영역
    # 초기에는 검색 결과 출력창을 숨김 (visible=False)
    retrieved_docs_display = gr.Textbox(label="검색된 문서들", lines=10, visible=False)
    msg = gr.Textbox(label="질문해주세요!")                # 사용자 질문 입력창
    clear = gr.Button("대화 초기화")                       # 대화 기록 초기화 버튼

    def respond(message, chat_history):
        result = chain_with_chat_history.invoke(
            {"question": message},
            config={"configurable": {"session_id": session_id}}
        )
        
        docs = cross_encoder_retriever(message)  # 문자열 리스트 반환
        docs_str = docs
        
        chat_history.append((message, result))
    
        if "검색 기반" in result:
            docs_update = gr.update(value=docs_str, visible=True)
        else:
            docs_update = gr.update(value="", visible=False)
    
        return "", chat_history, docs_update
        

    msg.submit(respond, [msg, chatbot], [msg, chatbot, retrieved_docs_display])
    clear.click(lambda: None, None, chatbot, queue=False)

demo.launch(debug=True)
```
