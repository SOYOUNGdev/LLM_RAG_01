```python
import openai

client = openai.OpenAI(api_key="")
```

```python
system_prompt = '''당신은 주어진 상담 대화로부터 고객의 개인 정보를 익명화하는 익명화봇입니다.
상담 대화는 상담사와 고객의 대화로 이루어져 있으며, 고객의 개인 정보를 보호하기 위해 아래의 기준에 따라 익명화해야 합니다.

### 익명화 기준
- "연락처": "[CONTACT1]", "[CONTACT2]" (고객이 여러 개의 연락처를 제공할 경우 구분)
- "주소": "[ADDRESS]"
- "사람 이름": "[PERSON1]", "[PERSON2]" (동일 인물은 같은 숫자로 유지, 다른 인물은 다른 숫자로 구분)
  - 상담사의 이름도 익명화합니다. 단, 상담사의 성만 언급된 경우에는 익명화 하지 않습니다. (예: "김상담사", "박상담사")
  - 고객의 이름이 다른 표현으로 등장해도 동일한 "[PERSONX]"로 표시합니다. (예: "홍길동 고객님", "길동 씨", "홍 고객님" → "[PERSON1]")
- "생년월일": "[DATE]"
  - 숫자로만 표현된 경우도 익명화합니다. (예: "800510" → "[DATE]")
- "주문번호": "[ORDERNUMBER1]", "[ORDERNUMBER2]" (여러 개 제공 시 구분)
- "계좌 번호": "[ACCOUNT1]", "[ACCOUNT2]" (여러 개 제공 시 구분)
- "카드 번호": "[ACCOUNT]"
- "유효 기간": "[EXPIRYDATE]"
- 숫자로 된 개인 정보(주문번호, 생년월일, 카드번호 등)는 다양한 형식(띄어쓰기, "-", "." 포함)으로 입력될 수 있으므로 모두 익명화합니다.
'''
```

```python
def get_results(user_prompt):
  response = client.chat.completions.create(
    model="gpt-4o",
    messages=[
          {"role": "system", "content": system_prompt},
          {"role": "user", "content": user_prompt}
      ],
    temperature=0
  )
  return response.choices[0].message.content
```

```python
user_prompt = """상담사: 안녕하세요 고객님, 저는 김상담사입니다. 오늘 어떤 문제로 도와드릴까요?

고객: 안녕하세요, 제 이름은 홍길동입니다. 전화번호는 010-1234-5678이고, 생년월일은 1980년 5월 10일이에요.

상담사: 네, 홍길동 고객님. 주소를 말씀해 주시겠어요?

고객: 서울시 강남구 테헤란로 123번지 456아파트 789호입니다.

상담사: 카드 결제를 원하시면 카드번호와 유효기간을 알려주세요.

고객: 카드 번호는 1234-5678-9012-3456이고, 유효기간은 2025년 10월입니다.
"""
    
anonymized_text = get_results(user_prompt)
print(anonymized_text)
```
